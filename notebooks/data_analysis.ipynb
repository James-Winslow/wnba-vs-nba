```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the combined data
data = pd.read_csv('../data/combined_data.csv')

# Setting up the plot style using seaborn for better aesthetics
sns.set(style="whitegrid")
plt.figure(figsize=(16, 10))

# Custom color palette
palette = {"WNBA": "skyblue", "NBA": "lightcoral"}

# Box plots for shooting percentages with enhanced visuals
plt.subplot(1, 3, 1)
sns.boxplot(x='League', y='FG%', data=data, palette=palette)
sns.stripplot(x='League', y='FG%', data=data, color='black', size=4, jitter=True, alpha=0.5)
plt.title('Field Goal Percentage', fontsize=14)
plt.xlabel('League', fontsize=12)
plt.ylabel('FG%', fontsize=12)

plt.subplot(1, 3, 2)
sns.boxplot(x='League', y='FT%', data=data, palette=palette)
sns.stripplot(x='League', y='FT%', data=data, color='black', size=4, jitter=True, alpha=0.5)
plt.title('Free Throw Percentage', fontsize=14)
plt.xlabel('League', fontsize=12)
plt.ylabel('FT%', fontsize=12)

plt.subplot(1, 3, 3)
sns.boxplot(x='League', y='3P%', data=data, palette=palette)
sns.stripplot(x='League', y='3P%', data=data, color='black', size=4, jitter=True, alpha=0.5)
plt.title('Three Point Percentage', fontsize=14)
plt.xlabel('League', fontsize=12)
plt.ylabel('3P%', fontsize=12)

# Adding a main title and a subtitle
plt.suptitle('Shooting Accuracy Comparison Between WNBA and NBA Players', fontsize=16)
plt.subplots_adjust(top=0.85, wspace=0.3)

# Display the plot
plt.show()
```
